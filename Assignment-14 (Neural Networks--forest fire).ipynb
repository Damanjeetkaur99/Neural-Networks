### X for input and Y for output var

```python
x=df_ff.iloc[:,0:28]
y=df_ff.iloc[:,29]
```

```python
x
```

```python
y
```

```python
a = StandardScaler()
a.fit(x)
X_standardized = a.transform(x)
```

```python
pd.DataFrame(X_standardized).describe()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

#### creating model with epochs=200 and batch_size=40

```python
# create model
model = Sequential()
model.add(Dense(30, input_dim=28,  activation='relu')) #1st layer dense is neurons and input dim is dimensions of input
model.add(Dense(28,  activation='relu')) #2nd layer or hidden
model.add(Dense(1,  activation='sigmoid')) #3rd layer or output layer where only 1 output 
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, validation_split=0.33,epochs=200, batch_size=40) #epoch means no. of iterations more of it more accuracy
```

```python
scores = model.evaluate(x, y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

#### creating model with epochs=150 and batch_size=30

```python
model.fit(x, y, validation_split=0.33,epochs=150, batch_size=30) #epoch means no. of iterations more of it more accuracy
```

```python
scores = model.evaluate(x, y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
